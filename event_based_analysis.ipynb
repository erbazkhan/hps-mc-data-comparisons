```python
f = R.TFile.Open("/home/ekhan/mc_data_comparison/montecarlo/tritrig_pulser/tritrig_pulser_101_to_110.minidst")
print("Is file open?", f and not f.IsZombie())
f.ls()

# Get the tree
mc_tree = f.Get("MiniDST")
```

```python
evt = 5
mc_tree.GetEntry(evt)

print("Nclusters:", len(mc_tree.ecal_cluster_energy))
print("ecal_cluster_energy:", list(mc_tree.ecal_cluster_energy))
print("ecal_cluster_mc_id:", list(mc_tree.ecal_cluster_mc_id))

print("Nparts:", len(mc_tree.part_pz))
print("part_ecal_cluster:", list(mc_tree.part_ecal_cluster))
print("part_pz:", list(mc_tree.part_pz))
```

```python
evt = 5
mc_tree.GetEntry(evt)

print("Clusters:")
for c in range(len(mc_tree.ecal_cluster_energy)):
    print(f"  c{c}: E={mc_tree.ecal_cluster_energy[c]:.3f} mc_id={int(mc_tree.ecal_cluster_mc_id[c])}")

print("\nParts:")
for i in range(len(mc_tree.part_ecal_cluster)):
    c = int(mc_tree.part_ecal_cluster[i])
    if c >= 0:
        mcid = int(mc_tree.ecal_cluster_mc_id[c])
        print(f"  p{i}: pz={mc_tree.part_pz[i]:.3f}  clus={c}  clE={mc_tree.ecal_cluster_energy[c]:.3f}  mcid={mcid}")
    else:
        print(f"  p{i}: pz={mc_tree.part_pz[i]:.3f}  clus=-1 (track-only)")
```
